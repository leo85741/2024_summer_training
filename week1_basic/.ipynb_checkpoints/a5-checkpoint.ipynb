```python
import requests
from bs4 import BeautifulSoup
import json

# pchome茶葉類商品
# 內容至少包含商品名稱、商品描述、商品價格、原始價格、連結
url = "https://24h.pchome.com.tw/search/v4.3/all/results?q=%E8%8C%B6%E8%91%89&page=2&sort=rnk/dc"

response = requests.get(url)
response.encoding = 'utf-8'  # 設置編碼

data = json.loads(response.text)

# 提取商品資訊
for item in data['Prods']:
    商品名稱 = item['Name']
    品牌 = item['Brand'] if item['Brand'] else '無'
    說明 = item['Describe']
    原始價格 = item['OriginPrice']
    價格 = item['Price']
    連結 = f"https://24h.pchome.com.tw/prod/{item['Id']}"

    # 顯示
    print(f"商品名稱: {商品名稱}")
    print(f"品牌: {品牌}")
    print(f"說明: {說明}")
    print(f"原始價格: {原始價格}")
    print(f"價格: {價格}")
    print(f"連結: {連結}")
    print("-" * 30)
```
