## 3. lexicon-base 情緒分析
進行基於情緒字典的情緒分析，首先需要準備情緒字典後再將字典與斷詞後的文章進行合併

### 3.1 準備LIWC情緒字典與清理完成的傲慢與偏見
- 全名Linguistic Inquiry and Word Counts，由心理學家Pennebaker於2001出版  
- 分為正向情緒與負向情緒

```python
# 讀取情緒字典
def read(filepath):
    with open(filepath, 'r') as file:
        return file.read()
    
    
positive_words = read('./dict/liwc/en/positive.txt').split('\n')
negative_words = read('./dict/liwc/en/negative.txt').split('\n')

df_positive_words = pd.DataFrame({'word': positive_words, 'sentiment': 'positive'})
df_negative_words = pd.DataFrame({'word': negative_words, 'sentiment': 'negative'})

# 合併 DataFrame
df_words = pd.concat([df_negative_words, df_positive_words], ignore_index=True)
df_words
```

```python
## 讀入清理過的傲慢與偏見
pride = read_data('./pride_word.csv')
pride
```

### 3.2 情緒字典與斷詞結果進行合併
將情緒字典與斷詞結果進行合併。

```python
# pride_lemma對照df_words_word做合併

# 用 pandas中的 merge()，做left_join
merge = pd.merge(pride, df_words, left_on='lemma', right_on='word', how='left')

# 移除多餘的 word 欄位
merge = merge.drop(columns=['word_y'])
merge.rename(columns={'word_x':'word'})
merge
```

```python
# 算出每章情緒總和(sentiment_count)
sentiment_count = merge.groupby(['chapter','sentiment']).size().reset_index(name='size')

new = {'chapter':0,'sentiment':'positive','size':0}
sentiment_count = pd.DataFrame([new]).append(sentiment_count,ignore_index=True)

sentiment_count
```

### 3.3 正負向情緒分數聲量長條
分別畫出文集中，正向（positive）與負向（negative）情緒的折線圖

```python
posi = sentiment_count[sentiment_count['sentiment'] == 'positive']
nega = sentiment_count[sentiment_count['sentiment'] == 'negative']

# 重置index
posi = posi.reset_index(drop=True)
nega = nega.reset_index(drop=True)
print(posi)
print(nega)

plt.figure(figsize=(10, 6))

# bar:垂直長條圖 barh:水平長條圖 label:圖例
plt.bar(posi['chapter'], posi['size'], color='blue',label='positive', alpha=0.6)
plt.bar(nega['chapter'], nega['size'], color='orange',label='negative', alpha=0.6, bottom=posi['size'])

plt.xlabel('Chapter')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 排序
posi_sort = posi.sort_values(by='size',ascending=False)
posi_sort
```

畫出每章的情緒總分數，可以看到正向情緒在 18 與 43 章較高

### 3.4 正負向情緒分數比例折線圖圖

```python
# 計算各章節情緒總數
sum = sentiment_count.groupby('chapter')['size'].transform('sum')
sentiment_count['ratio'] = sentiment_count['size']/sum
sentiment_count
```

```python
plt.figure(figsize=(10, 6))

# plot:折線圖
plt.plot(posi['chapter'], posi['size'], color='blue', marker='', label='positive')
plt.plot(nega['chapter'], nega['size'], color='orange', marker='', label='negative')


plt.xlabel('Chapter')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()
```

### 3.5 情緒代表字
從整個文集中計算出正負向情緒字的出現詞頻，並找出情緒的代表字

```python
# 計算整個文集的詞頻
sentiments_count = merge.groupby(['word_x']).size().reset_index(name='size')

# 與情緒字典合併
ans_merge = pd.merge(sentiments_count, df_words, left_on='word_x', right_on='word', how='left')
ans_merge = ans_merge.drop(columns=['word'])

# 保留情緒字眼並排序
ans = ans_merge[(ans_merge['sentiment'] == 'negative') | (ans_merge['sentiment'] == 'positive')]
ans = ans.sort_values(by='size',ascending=False)

# 重新排列
ans = ans[['word_x', 'sentiment', 'size']]

ans
```

```python
posi = ans[ans['sentiment'] == 'positive'].head(5)
nega = ans[ans['sentiment'] == 'negative'].head(5)

posi = posi.sort_values(by='size', ascending=True)
nega = nega.sort_values(by='size', ascending=True)


# bar:垂直長條圖 barh:水平長條圖 label:圖例
# POSITIVE
plt.barh(posi['word_x'], posi['size'], color='blue',label='positive', alpha=0.7)

plt.xlabel('Word')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()

# NEGATIVE
plt.barh(nega['word_x'], nega['size'], color='orange',label='negative', alpha=0.7)
plt.xlabel('Word')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()
```

## 3.6 歸類正負面句子
- 之前的情緒分析大部分是章節的詞彙全部加總，接下來將正負面情緒的句子分開，看看能不能發現一些新的東西。  
- 接下來歸類章節，將每一章句子正負面情緒的分數算出來，然後大概分類句子屬於正面還是負面。

```python
merge
```

```python
# 算出每章情緒總和(sentiment_count)
# 抓取有情緒的字
merged = merge[(merge['sentiment'] == 'negative')|(merge['sentiment'] == 'positive')]

# 合併
sentiment_count = merge.groupby(['sentence', 'chapter', 'sentiment']).size().reset_index(name='size')

sentiment_count
```

```python
# 合併
sentiment_m = sentiment_count.pivot_table(index='sentence', columns='sentiment', values='size', fill_value=0).reset_index()
sentiment_m['negative']=sentiment_m['negative'].astype(int)
sentiment_m['positive']=sentiment_m['positive'].astype(int)
sentiment_m
```

```python
# 如果該句子的正面情緒大於負面情緒，則歸類該句子為正面(positive)，反之則歸類為負面(negative)
# 用 numpy的 where函數
sentiment_m['type'] = np.where(sentiment_m['positive'] > sentiment_m['negative'], 'positive', 'negative')
sentiment_m
```

```python
result = sentiment_m.groupby(['type']).size().reset_index(name='size')
result
```

### 正負情緒句子數量統計圖

```python
# 合併(相較前面多了Chapter欄位)
sentiment_c = sentiment_count.pivot_table(index=['chapter','sentence'], columns='sentiment', values='size', fill_value=0).reset_index()
sentiment_c['negative']=sentiment_c['negative'].astype(int)
sentiment_c['positive']=sentiment_c['positive'].astype(int)
sentiment_c['type'] = np.where(sentiment_c['positive'] > sentiment_c['negative'], 'positive', 'negative')

# 重新排列
sentiment_c = sentiment_c[['sentence','negative','positive','type','chapter']]

sentiment_c
```

```python
# 對 'chapter','type' 做合併
result = sentiment_c.groupby(['chapter','type']).size().reset_index(name='size')
result
```

```python
posi = result[result['type'] == 'positive']
nega = result[result['type'] == 'negative']

# 重置index
posi = posi.reset_index(drop=True)
nega = nega.reset_index(drop=True)
# print(posi)
# print(nega)


# bar:垂直長條圖 barh:水平長條圖 label:圖例
plt.bar(posi['chapter'], posi['size'], color='blue',label='positive', alpha=0.6)


plt.xlabel('Chapter')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()

plt.bar(nega['chapter'], nega['size'], color='orange',label='negative', alpha=0.6)
plt.xlabel('Chapter')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()
```
