## 3. lexicon-base 情緒分析
進行基於情緒字典的情緒分析，首先需要準備情緒字典後再將字典與斷詞後的文章進行合併

## 3.1 準備LIWC字典  
- 全名Linguistic Inquiry and Word Counts，由心理學家Pennebaker於2001出版  
- 分為正向情緒與負向情緒

```python
# 讀取情緒字典
```

```python
# 讀取情緒字典
def read(filepath):
    with open(filepath, 'r') as file:
        return file.read()
    
    
positive_words = read('./dict/liwc/positive.txt').split(',')
negative_words = read('./dict/liwc/negative.txt').split(',')

df_positive_words = pd.DataFrame({'word': positive_words, 'sentiment': 'positive'})
df_negative_words = pd.DataFrame({'word': negative_words, 'sentiment': 'negative'})

# 合併 DataFrame
df_words = pd.concat([df_negative_words, df_positive_words], ignore_index=True)
df_words
```

```python
## 讀入清理過的  ptt
ptt_word = read_data('./ptt_word.csv')
ptt_word
```

### 3.2 情緒字典與斷詞結果進行合併
將情緒字典與斷詞結果進行合併。。

```python
# 用 pandas中的 merge()，做left_join
merge = pd.merge(ptt_word, df_words, on='word', how='left')
merge = merge.drop(columns=['artTitle','artTime','artPoster','artCat','commentNum','push','boo'])
merge['artDate'] = pd.to_datetime(merge['artDate'],format='%Y-%m-%d')
merge
```

### 3.3 正負向情緒分數聲量折線圖圖
分別畫出文集中，正向（positive）與負向（negative）情緒的折線圖

```python
merge = merge[(merge['sentiment'] == 'negative')|(merge['sentiment'] == 'positive')]
sentiments_count = merge.groupby(['artDate','sentiment']).size().reset_index(name='size')

sentiments_count
```

```python
posi = sentiments_count[sentiments_count['sentiment'] == 'positive']
nega = sentiments_count[sentiments_count['sentiment'] == 'negative']

# 重置index
posi = posi.reset_index(drop=True)
nega = nega.reset_index(drop=True)

# plot:折線圖
plt.plot(posi['artDate'], posi['size'], color='orange', marker='', label='positive')
plt.plot(nega['artDate'], nega['size'], color='blue', marker='', label='negative')


plt.xlabel('artDate')
plt.ylabel('Size')
plt.legend()

plt.tight_layout()
plt.show()
```

畫出每天的情緒總分數，可以看到大概在3/18後，短短的幾天內，情緒從正面為主轉為負面為主。約在20號之後討論度逐漸下降

### 3.4 正負向情緒分數比例折線圖圖

```python
# 計算各章節情緒總數
sum = sentiments_count.groupby('artDate')['size'].transform('sum')
sentiments_count['ratio'] = sentiments_count['size']/sum
sentiments_count
```

```python
posi = sentiments_count[sentiments_count['sentiment'] == 'positive']
nega = sentiments_count[sentiments_count['sentiment'] == 'negative']

# 重置index
posi = posi.reset_index(drop=True)
nega = nega.reset_index(drop=True)

# plot:折線圖
plt.plot(posi['artDate'], posi['ratio'], color='orange', marker='', label='positive')
plt.plot(nega['artDate'], nega['ratio'], color='blue', marker='', label='negative')


plt.xlabel('artDate')
plt.ylabel('ratio')
plt.legend()

plt.tight_layout()
plt.show()
```

透過上圖我們可以發現3/18到3/19是一次情緒由正轉負的轉折，而3/25到3/26是一次情緒由負轉正的轉折

### 3.4 文字雲
針對不同日期畫出該天的文字雲，如果文字雲中的字詞大小越大，代表出現頻率越高

### 2021-03-17 文字雲
2021-03-17的文字雲，往前看17正面情緒較高的文字雲，發現此時負面批評詞彙較少，出現較多正面詞彙或中性詞彙，如「笑死」、「年輕人」、「行銷」、「活動」、「廣告」，推測此時負面文章較少，網友對鮭魚事件仍持一種較為幽默的態度，覺得此企劃好笑、年輕人太衝動等。

```python
word_count = merge.groupby(['artDate','word']).size().reset_index(name='size')
word_count = word_count.sort_values(by='size', ascending=False)
word_count
```

```python
word_count
```

```python
# wordcloud 的 input 是 dictionary
# 篩選出 2021-3-17
top_words =word_count[word_count['artDate']=='2021-03-17'].head(25)
top_words

# 創建詞頻字典
word_freq_dict = dict(zip(top_words['word'], top_words['size']))
word_freq_dict
```

```python
font = './SourceHanSansTW-Regular.otf' #中文字型路徑
    
    
# 創建文字雲
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=font).generate_from_frequencies(word_freq_dict)
    
# 繪製文字雲
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'2021/3/17')
plt.show()
```

### 2021-03-19 文字雲
先從2021-03-19的情緒高點看起，呼應上面負面的情緒分析，出現「浪費」、「貪小便宜」、「丟臉」、「乞丐」等詞彙。推測是因許多網友抨擊改名行為不明智，也批評部分免費吃壽司的人造成食物浪費。  
而「新聞」、「日本」、「韓國」等詞，則是討論台灣鮭魚改名之亂引起國際媒體關注


```python
# 0319
top_words =word_count[word_count['artDate']=='2021-03-19'].head(25)
top_words

# 創建詞頻字典
word_freq_dict = dict(zip(top_words['word'], top_words['size']))
word_freq_dict
```

```python
# 創建文字雲
wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=font).generate_from_frequencies(word_freq_dict)
    
# 繪製文字雲
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'2021/3/19')
plt.show()
```

### 3.5 情緒代表字
從整個文集中計算出正負向情緒字的出現詞頻，並找出情緒的代表字

```python
sentiments_count = merge.groupby(['word','sentiment']).size().reset_index(name='size')
sentiments_count = sentiments_count.sort_values(by='size', ascending=False)
sentiments_count
```

```python
# 正向情緒df
posi = sentiments_count[sentiments_count['sentiment'] == 'positive']
posi
```

```python
# 負向情緒df
nega = sentiments_count[sentiments_count['sentiment'] == 'negative']
nega
```

```python
plt.rcParams['font.sans-serif'] = ['Noto Sans CJK JP'] # 設定中文字型, 參考：https://pyecontech.com/2020/03/27/python_matplotlib_chinese/

posi = posi.head(15)
nega = nega.head(15)

posi = posi.sort_values(by='size', ascending=True)
nega = nega.sort_values(by='size', ascending=True)

# POSITIVE
plt.barh(posi['word'], posi['size'], color='orange',label='positive', alpha=0.7)

plt.xlabel('Size')
plt.ylabel('Word')
plt.legend()

plt.tight_layout()
plt.show()

# NEGATIVE
plt.barh(nega['word'], nega['size'], color='blue',label='negative', alpha=0.7)
plt.xlabel('Size')
plt.ylabel('Word')
plt.legend()

plt.tight_layout()
plt.show()
```

另外，也可以依據不同日期觀察情緒代表字的變化
### 2021-03-17 正負情緒代表字

```python
#篩選日期 2021-03-17
merge_17 = merge[merge['artDate']=='2021-03-17']
sentiments_count = merge_17.groupby(['artDate','word','sentiment']).size().reset_index(name='size')
sentiments_count = sentiments_count.sort_values(by='size', ascending=False)
sentiments_count
```

```python
#篩選日期 2021-03-17
posi = sentiments_count[sentiments_count['sentiment'] == 'positive'].head(15)
nega = sentiments_count[sentiments_count['sentiment'] == 'negative'].head(15).head(15)

posi = posi.sort_values(by='size', ascending=True)
nega = nega.sort_values(by='size', ascending=True)

# POSITIVE
plt.barh(posi['word'], posi['size'], color='orange',label='positive', alpha=0.7)

plt.xlabel('Size')
plt.ylabel('Word')
plt.legend()

plt.tight_layout()
plt.show()

# NEGATIVE
plt.barh(nega['word'], nega['size'], color='blue',label='negative', alpha=0.7)
plt.xlabel('Size')
plt.ylabel('Word')
plt.legend()

plt.tight_layout()
plt.show()
```
