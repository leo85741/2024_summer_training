## 讀入所需的libary

```python
import pandas as pd
import numpy as np 
import nltk
import math
nltk.download("punkt") # 下載需要用到的語料庫
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

#### 讀入PRIDE_PREJUDICE 資料

```python
pride = pd.read_csv("./pride_word.csv")[['sentence','chapter','word']]
pride.head()
```

# tfidf
tfidf是一種文字分析方式，透過計算文字在文章中和整篇文本中出現的次數來給予權重，來計算該字對於文本的重要性

## 計算詞頻(term frequency)

### 計算某文字在各chapter中出現次數

```python
freq = pride.groupby(['chapter','word']).size().reset_index(name='chapter_word')
freq = freq.sort_values(by='chapter_word',ascending=False)
freq
```

### 計算每個文章中總共有多少字


```python
chapter_freq = freq.groupby(['chapter']).size().reset_index(name='chapter_total')
chapter_freq
```

### 算tf
在文章中出現的比例，出現越多次代表該詞對該文章有較大影響

```python
tf = pd.merge(freq, chapter_freq, on='chapter', how='left')
tf['tf'] = tf['chapter_word']/tf['chapter_total']
tf


```

### 算IDF
計算某詞出現在幾篇chapter中(就算某詞在一個文章中出現多次，IDF仍只算一次)，因為若某詞在所有文章都出現過，代表該詞的資訊量較少

```python
# 章節數
chapter_size = 61

idf = freq.groupby(['word']).size().reset_index(name='size')

idf['idf'] = chapter_size/idf['size']
idf['idf'] = idf['idf'].apply(math.log)            # 用 pandas提供的方法取 log
idf = idf.sort_values(by='idf',ascending=False)    # 排序
idf
```

### 算tfidf
將tf 和 idf 相乘即得出tfidf

```python
tfidf = pd.merge(tf, idf, on='word', how='left')
tfidf = tfidf.drop(columns=['size'])

tfidf['tfidf'] = tfidf['tf']*tfidf['idf']
tfidf
```

### 由小到大檢視tfidf

```python
tfidf = tfidf.sort_values(by='tfidf',ascending=True)    
tfidf
```

#### 找出tfidf較高的字

先找出每個chapter前10名的TFIDF

```python
# 對 Chapter做群組，使用 nlargest找出前10名
top_tfidf = tfidf.groupby('chapter').apply(lambda x: x.nlargest(10,'tfidf')).reset_index(drop=True)
top_tfidf
```

在統計每個字出現在每篇tfidf前10名的次數

```python
top_freq = top_tfidf.groupby(['word']).size().reset_index(name='size')
top_freq = top_freq.sort_values(by='size',ascending=False)    

top_freq
```

# ngram

### bigram

```python
# 只保留不重複的句子
bigram = pride[['sentence']].drop_duplicates().reset_index(drop=True)
bigram['sentence'] = bigram['sentence'].astype(str)


#用 bigram套件進行拆分
from nltk import bigrams

def generate_bigrams(sentence):
    tokens = nltk.word_tokenize(sentence)
    bigram_list = list(bigrams(tokens))
    
    return bigram_list

bigram['bigrams'] = bigram['sentence'].apply(generate_bigrams)

bigram = bigram.explode('bigrams').reset_index(drop=True)
bigram = bigram.drop(columns='sentence')
bigram
```

### trigram

```python
trigram = pride[['sentence']].drop_duplicates().reset_index(drop=True)
trigram['sentence'] = trigram['sentence'].astype(str)

#用 bigram套件進行拆分
from nltk import trigrams

def generate_trigrams(sentence):
    tokens = nltk.word_tokenize(sentence)
    trigram_list = list(trigrams(tokens))
    return trigram_list

trigram['trigrams'] = trigram['sentence'].apply(generate_trigrams)

trigram = trigram.explode('trigrams').reset_index(drop=True)
trigram = trigram.drop(columns='sentence')
trigram
    
```

### bigram 次數

```python
bigram_count = bigram.groupby(['bigrams']).size().reset_index(name='size')
bigram_count = bigram_count.sort_values(by='size',ascending=False)     

bigram_count
```

拆成兩個columns 分別為word1 word2

```python
bigram_word = pd.DataFrame()
bigram_word[['word1','word2']] = pd.DataFrame(bigram['bigrams'].tolist(), index=bigram.index)
bigram_word
```

計算street跟其他自一起出現的頻率

```python
# 找出含有'street'的 bigram
bi_street = bigram_word[(bigram_word['word1']=='street') | (bigram_word['word2']=='street')]

# 取出單字
other = pd.DataFrame()
other['word'] = bi_street.apply(lambda row:row['word1'] if row['word2']=='street' else row['word2'], axis=1)

# 計算頻率
other = other.groupby(['word']).size().reset_index(name='freq')
other = other.sort_values(by='freq',ascending=False)
other
```

### 計算每個字的correlation


#### 轉成DTM格式

```python
pride['word'] = pride['word'].astype(str)
table = pride.groupby('chapter')['word'].apply(lambda x: ' '.join(x)).reset_index()
table
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(table['word'])
result = pd.DataFrame(x.toarray(), columns=vectorizer.get_feature_names(), index=table['chapter'])
result
```

#### 計算每個字對每篇文章的pearson correlation
hint:pandas有提供function，但是算得很慢，可以用numpy的

```python
# correlation_ma = np.corrcoef(result.values, rowvar=False)
# correlation_ma

# columns = result.columns
# index =  result.columns

# correlation_df = pd.DataFrame(correlation_ma, index=index, columns=columns)
# correlation_df

correlation_df = result.corr(method='pearson')

correlation_df
```

```python
correlation_df.reset_index(drop=True,inplace=True)
re_index = correlation_df.columns.tolist()
correlation_df.insert(0, 'item1', re_index)

correlation_df
```

轉成一對一的格式後，把corr是1的去掉

```python
correlation_index = correlation_df.melt(id_vars=['item1'], var_name='item2', value_name='correlation')
correlation_index = correlation_index[correlation_index['correlation']!= 1]
correlation_index
```

#### 列出跟 elizabeth 還有my最相關的前6個字

```python
# 'elizabeth'最相關的前6個字
ezli_top = correlation_index[correlation_index['item2']=='elizabeth'].nlargest(6,'correlation')

ezli_top = ezli_top.sort_values(by='correlation', ascending=True)

plt.barh(ezli_top['item1'], ezli_top['correlation'], color='blue', alpha=0.7)

plt.xlabel('correlation')
plt.ylabel('item1')
plt.title('elizabeth')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 'pride'最相關的前6個字
pride_top = correlation_index[correlation_index['item2']=='pride'].nlargest(6,'correlation')

pride_top = pride_top.sort_values(by='correlation', ascending=True)

plt.barh(pride_top['item1'], pride_top['correlation'], color='blue', alpha=0.7)

plt.xlabel('correlation')
plt.ylabel('item1')
plt.title('pride')
plt.legend()

plt.tight_layout()
plt.show()
```
